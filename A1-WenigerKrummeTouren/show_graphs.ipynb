```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
coords = []
with open(f'beispieldaten/wenigerkrumm{input("number: > ")}.txt') as f:
    for line in f:
        coords.append(tuple(map(float, line.split())))

G = nx.Graph()
G.add_nodes_from(range(len(coords)))
nx.draw(G, coords)
```
